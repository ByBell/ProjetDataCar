```python
import pandas as pd
from sklearn.dummy import DummyClassifier
from sklearn.metrics import precision_recall_fscore_support, precision_score, recall_score, f1_score
```

```python
file = 'winequality-white.csv'

df = pd.read_csv(file, sep=';')
X_COL = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
y_COL = ['<5', '<6', '<7'] #['quality']

df.loc[df['quality'] < 5, '<5'] = True
df.loc[df['quality'] < 6, '<6'] = True
df.loc[df['quality'] < 7, '<7'] = True
df.fillna(False, inplace=True)

df[df['<7'] == False]
```

```python
def model_score(model, dataframeTrain, dataframeTest, X_col_array, y_col_array):

    for idx, col in enumerate(y_col_array):
        dfX, dfy = dataframeTrain[X_col_array], dataframeTrain[col]
        
        model.fit(dfX, dfy)

        pred = model.predict(dataframeTest[X_col_array])

        print(col, precision_recall_fscore_support(dataframeTest[col], pred, average="weighted"))
```

```python
dfTrain = df[-500:]
dfTest = df[0:-500]
```

```python
dummy_clf_mf = DummyClassifier(strategy="most_frequent")
model_score(dummy_clf_mf, dfTrain, dfTest, X_COL, y_COL)
```

```python
dummy_clf_mf = DummyClassifier(random_state=5)
model_score(dummy_clf_mf, dfTrain, dfTest, X_COL, y_COL)
```

```python
dummy_clf_mf = DummyClassifier(strategy="uniform")
model_score(dummy_clf_mf, dfTrain, dfTest, X_COL, y_COL)
```
