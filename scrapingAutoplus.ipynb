```python
def scrap_page(page_number, session):
    df = pd.DataFrame(columns = ['MARQUE_MODELE', 'VERSION', 'ENERGIE', 'BOITE_VITESSE', 'PUISS_FISCALE', 'DATE', 'MALUS', 'PRIX_NEUF'])
    
    try :
        s = session.get(url = 'https://www.autoplus.fr/prix-neuf/?page=' + str(page_number))
    except urllib3.exceptions.ProtocolError:
        print("Error RemoteDisconnected")
        time.sleep(np.random.ranf()*10)
        s = session.get(url = 'https://www.autoplus.fr/prix-neuf/?page=' + str(page_number))
        
    for i in range(1, 32):
        elements_list = s.html.xpath('//*[@id="content"]/div[6]/table/tbody/tr['+str(i)+']')[0].text.split('\n')
        elements_list = [elem.replace(u'\xa0', ''). replace('€', '') for elem in elements_list]
        
        try :
            df.loc[i] = elements_list
        except ValueError:
            pass
        
    return df
```

```python
#df = pd.DataFrame(columns = ['MARQUE_MODELE', 'VERSION', 'ENERGIE', 'BOITE_VITESSE', 'PUISS_FISCALE', 'DATE', 'MALUS', 'PRIX_NEUF'])
df = pd.read_csv('autoplus.csv')
session = HTMLSession()

for i in range(109, 460):
    session = HTMLSession()            
    newDf = scrap_page(i, session)
        
    df = pd.concat([df, newDf])
    df.to_csv('autoplus.csv')

    random_wait_time = np.random.ranf()*3
    print('page', i, 'fetchée... attente de', random_wait_time, 's ...')
    time.sleep(random_wait_time)
```
